```python
import matplotlib.pyplot as plt

# Данные
days = ['Пн', 'Вт', 'Ср', 'Чт', 'Пт', 'Сб', 'Вс']
temperature = [3, 5, 7, 10, 12, 8, 6]

# Создание графика
plt.figure(figsize=(10, 6))
plt.plot(days, temperature, marker='o', linestyle='-', color='crimson', linewidth=2)

# Настройка оформления
plt.title('Изменение температуры воздуха за неделю', fontsize=14, pad=20)
plt.xlabel('Дни недели', fontsize=12)
plt.ylabel('Температура (°C)', fontsize=12)
plt.grid(True, alpha=0.4, linestyle='--')

# Подписи значений
for x, y in zip(days, temperature):
    plt.text(x, y+0.3, f'{y}°C', ha='center', va='bottom', fontsize=10)

plt.ylim(0, 15)  # Установка диапазона оси Y
plt.tight_layout()
plt.show()
```

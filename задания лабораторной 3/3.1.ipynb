```python
import matplotlib.pyplot as plt
import numpy as np

# Задаем интервал значений x
x = np.linspace(-10, 10, 400)

# Вычисляем значения функции y = x^2
y = x**2

# Создаем график
plt.plot(x, y, label='y = x^2')

# Добавляем подписи осей
plt.xlabel('x')
plt.ylabel('y')

# Добавляем сетку
plt.grid(True)

# Добавляем заголовок
plt.title('График функции y = x^2')

# Добавляем легенду
plt.legend()

# Показываем график
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Создание данных
x = np.linspace(-2, 2, 400)
y1 = x
y2 = x**2
y3 = x**3

# Создание графика
plt.figure(figsize=(10, 6))

# Построение линий с разными стилями
plt.plot(x, y1, linestyle='--', color='blue', label='y = x')
plt.plot(x, y2, linestyle='-.', color='green', label='y = x²')
plt.plot(x, y3, linestyle='-', color='red', label='y = x³')

# Настройки оформления
plt.title('Графики функций')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True, alpha=0.3)
plt.axis('equal')  # Одинаковый масштаб осей
plt.legend()

# Автоматическая подгонка границ
plt.autoscale(enable=True, axis='both', tight=True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Генерация случайных данных
np.random.seed(42)  # Для воспроизводимости
x = np.random.rand(50) * 10  # X в диапазоне [0, 10]
y = np.random.rand(50) * 10  # Y в диапазоне [0, 10]

# Создание графика
plt.figure(figsize=(10, 6))
scatter = plt.scatter(
    x, y,
    c=x,                    # Цвет зависит от X
    s=y*50,                 # Размер зависит от Y (масштабируем)
    cmap='viridis',          # Цветовая карта
    alpha=0.7,              # Полупрозрачность
    edgecolors='black'      # Черные границы точек
)

# Настройка оформления
plt.title('Диаграмма рассеяния: цвет ~ X, размер ~ Y', pad=20)
plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.grid(alpha=0.3)

# Цветовая шкала для оси X
cbar = plt.colorbar(scatter)
cbar.set_label('Значение координаты X')

# Настройка легенды для размеров
sizes_legend = [2, 4, 6, 8, 10]
plt.legend(
    handles=[
        plt.Line2D([0], [0], marker='o', color='w', 
                  markersize=np.sqrt(size*50), 
                  markerfacecolor='gray', 
                  markeredgecolor='black') 
        for size in sizes_legend
    ],
    labels=[f'y = {size}' for size in sizes_legend],
    title='Размер точек ~ Y',
    loc='upper left'
)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Генерация данных
np.random.seed(42)  # Для воспроизводимости
data = np.random.normal(loc=0, scale=1, size=1000)

# Расчет среднего значения
mean = np.mean(data)

# Создание гистограммы
plt.figure(figsize=(10, 6))
n, bins, patches = plt.hist(
    data,
    bins=30,
    density=True,      # Нормировка площади к 1
    color="skyblue",
    edgecolor="white",
    alpha=0.7
)

# Вертикальная линия среднего
plt.axvline(
    x=mean,
    color="red",
    linestyle="--",
    linewidth=2,
    label=f"Среднее: {mean:.2f}"
)

# Настройка оформления
plt.title("Гистограмма нормального распределения (μ=0, σ=1)", pad=20)
plt.xlabel("Значение")
plt.ylabel("Плотность")
plt.grid(alpha=0.3, ls=":")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Данные
grades = ["Отлично", "Хорошо", "Удовлетворительно", "Неудовлетворительно"]
students = [20, 35, 30, 15]
colors = ['#4CAF50', '#2196F3', '#FFC107', '#F44336']  # Зеленый, синий, желтый, красный

# Создание графика
plt.figure(figsize=(10, 6))
bars = plt.bar(grades, students, color=colors, edgecolor='black')

# Настройка оформления
plt.title('Распределение оценок студентов', fontsize=14, pad=20)
plt.xlabel('Оценки', fontsize=12)
plt.ylabel('Количество студентов', fontsize=12)
plt.grid(axis='y', alpha=0.4, linestyle='--')

# Добавление значений над столбцами
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height}',
             ha='center', va='bottom')

plt.ylim(0, max(students) + 5)  # Запас по вертикали
plt.show()
```

```python
import matplotlib.pyplot as plt

# Данные из предыдущей задачи
grades = ["Отлично", "Хорошо", "Удовлетворительно", "Неудовлетворительно"]
students = [20, 35, 30, 15]
colors = ['#4CAF50', '#2196F3', '#FFC107', '#F44336']

# Расчет процентов
total = sum(students)
percentages = [(count/total)*100 for count in students]

# Построение круговой диаграммы
plt.figure(figsize=(8, 8))
wedges, texts, autotexts = plt.pie(
    students,
    labels=grades,
    colors=colors,
    autopct='%.1f%%',  # Формат подписей с процентами
    startangle=90,     # Начальный угол
    textprops={'fontsize': 12},
    wedgeprops={'edgecolor': 'black', 'linewidth': 0.5}
)

# Настройка подписей процентов
plt.setp(autotexts, color='white', weight='bold') 

# Добавление заголовка
plt.title('Распределение оценок студентов (в процентах)', fontsize=14, pad=20)

# Легенда (опционально)
plt.legend(
    wedges,
    [f'{l} ({p:.1f}%)' for l, p in zip(grades, percentages)],
    title="Оценки",
    loc="center left",
    bbox_to_anchor=(1, 0, 0.5, 1)
)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Создание сетки значений
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Вычисление Z
Z = np.sin(np.sqrt(X**2 + Y**2))

# Построение 3D-графика
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)

# Настройка оформления
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title(r'$z = \sin(\sqrt{x^2 + y^2})$')
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Создаем данные
x = np.linspace(-10, 10, 400)
x_trig = np.linspace(0, 2*np.pi, 400)

# Создаем сетку 2x2
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# График 1: Линейный
axs[0, 0].plot(x, x, color='blue')
axs[0, 0].set_title('Линейная функция: y = x')

# График 2: Парабола
axs[0, 1].plot(x, x**2, color='green')
axs[0, 1].set_title('Парабола: y = x²')

# График 3: Синус
axs[1, 0].plot(x_trig, np.sin(x_trig), color='red')
axs[1, 0].set_title('Синусоида: y = sin(x)')

# График 4: Косинус
axs[1, 1].plot(x_trig, np.cos(x_trig), color='purple')
axs[1, 1].set_title('Косинусоида: y = cos(x)')

# Общие настройки
for ax in axs.flat:
    ax.grid(alpha=0.4)
    ax.set_xlabel('x')
    ax.set_ylabel('y')

plt.tight_layout()
plt.show()
```

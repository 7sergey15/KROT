```python
import numpy as np
import matplotlib.pyplot as plt

# Вычисление аналитического интеграла
def f(x):
    return x**2

a, b = 0, 3
area = (b**3)/3 - (a**3)/3  # ∫x²dx = x³/3

# Генерация данных для графика
x = np.linspace(a, b, 400)
y = f(x)

# Создание графики
fig, ax = plt.subplots(figsize=(10, 6))

# Построение графика и заливка области
ax.plot(x, y, 'r-', linewidth=2)
ax.fill_between(x, y, alpha=0.3, color='blue')

# Настройка оформления
ax.set_title(r'Площадь под кривой $f(x) = x^2$ на [0, 3]', fontsize=14)
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('f(x)', fontsize=12)
ax.grid(True, linestyle='--', alpha=0.7)

# Добавление аннотации
ax.text(1.5, 5, f'Площадь = {area:.2f}', 
        fontsize=12, ha='center', 
        bbox=dict(facecolor='white', alpha=0.9))

# Настройка осей
ax.set_xlim([a-0.2, b+0.2])
ax.set_ylim([0, 10])

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Определение функции
def f(x):
    return np.sin(x)

# Вычисление аналитического интеграла
a, b = 0, np.pi
analytic_area = -np.cos(b) + np.cos(a)  # ∫sin(x)dx = -cos(x) + C

# Численная проверка
numeric_area, error = quad(f, a, b)

# Генерация данных для графика
x = np.linspace(a, b, 200)
y = f(x)

# Создание графики
fig, ax = plt.subplots(figsize=(10, 6))

# Построение графика и заливка области
ax.plot(x, y, 'r-', linewidth=2, label='f(x) = sin(x)')
ax.fill_between(x, y, alpha=0.3, color='blue')

# Настройка оформления
ax.set_title(r'Площадь под кривой $f(x) = \sin(x)$ на $[0, \pi]$', fontsize=14)
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('f(x)', fontsize=12)
ax.grid(True, linestyle='--', alpha=0.7)

# Отметки и аннотации
ax.set_xticks([0, np.pi/2, np.pi])
ax.set_xticklabels(['0', r'$\frac{\pi}{2}$', r'$\pi$'])
ax.text(np.pi/2, 0.4, f'Площадь = {analytic_area:.2f}', 
        fontsize=12, ha='center', va='center',
        bbox=dict(facecolor='white', alpha=0.9))

# Легенда и ограничения
ax.legend()
ax.set_xlim([a-0.2, b+0.2])
ax.set_ylim([0, 1.1])

plt.tight_layout()
plt.show()

# Вывод результатов
print(f"Аналитический результат: {analytic_area:.4f}")
print(f"Численный результат: {numeric_area:.4f}")
print(f"Погрешность: {error:.2e}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Определим функцию
def f(x):
    return 2 * np.log(x) + x

# Вычислим интеграл
area, error = quad(f, 1, 3)

# Создаём данные для графика
x = np.linspace(0.5, 4, 400)
y = f(x)

# Построим график
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(x, y, 'b', linewidth=2, label=r'$f(x) = 2\ln(x) + x$')
ax.set_ylim(bottom=0)

# Закрасим область под кривой на [1, 3]
ix = np.linspace(1, 3, 100)
iy = f(ix)
ax.fill_between(ix, iy, color='gray', alpha=0.3, label=f'Интеграл = {area:.4f}')

# Добавим вертикальные линии
ax.axvline(x=1, color='gray', linestyle='--', linewidth=1)
ax.axvline(x=3, color='gray', linestyle='--', linewidth=1)

# Оформление
ax.set_title(r'График $f(x) = 2\ln(x) + x$ и площадь на $[1, 3]$')
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.legend()
ax.grid(True)

# Заголовок с результатом
plt.title(f'$\\int_1^3 (2\\ln(x) + x)\\,dx = 6\\ln(3) \\approx {area:.4f}$')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Определим функцию
def f(x):
    return np.tan(x)

# Вычислим интеграл
area, error = quad(f, 0, np.pi/4)

# Создаём данные для графика
x = np.linspace(0, np.pi/4, 400)
y = f(x)

# Построим график
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(x, y, 'b', linewidth=2, label=r'$f(x) = \tan(x)$')
ax.set_ylim(bottom=0, top=1.1)

# Закрасим область под кривой
ix = np.linspace(0, np.pi/4, 100)
iy = f(ix)
ax.fill_between(ix, iy, color='gray', alpha=0.3, label=f'Интеграл = {area:.4f}')

# Добавим ось и оформление
ax.axhline(y=0, color='black', linewidth=0.5)
ax.set_title(r'График $f(x) = \tan(x)$ и площадь на $[0, \frac{\pi}{4}]$')
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.legend()
ax.grid(True)

# Корректная строка заголовка
title = r'$\int_0^{{\frac{{\pi}}{{4}}}} \tan(x)\,dx = \frac{{\ln(2)}}{{2}} \approx {:.4f}$'.format(area)
plt.title(title)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Определим функцию
def f(x):
    return x**2 + 1 / (x + 2)

# Вычислим интеграл
area, error = quad(f, 0, 3)

# Создаём данные для графика
x = np.linspace(-1, 4, 400)
y = f(x)

# Построим график
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(x, y, 'b', linewidth=2, label=r'$f(x) = x^2 + \frac{1}{x + 2}$')
ax.set_ylim(bottom=0, top=12)

# Закрасим область под кривой на [0, 3]
ix = np.linspace(0, 3, 100)
iy = f(ix)
ax.fill_between(ix, iy, color='gray', alpha=0.3, label=f'Площадь = {area:.4f}')

# Добавим вертикальные линии
ax.axvline(x=0, color='gray', linestyle='--', linewidth=1)
ax.axvline(x=3, color='gray', linestyle='--', linewidth=1)

# Оформление
ax.set_title(r'График $f(x) = x^2 + \frac{1}{x + 2}$ и площадь на $[0, 3]$')
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.legend()
ax.grid(True)

# Заголовок с результатом
plt.title(f'$\\int_0^3 \\left(x^2 + \\frac{{1}}{{x + 2}}\\right)dx = 9 + \\ln\\left(\\frac{{5}}{{2}}\\right) \\approx {area:.4f}$')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Определим функцию
def f(x):
    return (x**2) * np.exp(-x)

# Вычислим интеграл
area, error = quad(f, 0, 3)

# Создаём данные для графика
x = np.linspace(0, 4, 400)
y = f(x)

# Построим график
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(x, y, 'b', linewidth=2, label=r'$f(x) = x^2 e^{-x}$')
ax.set_ylim(bottom=0)

# Закрасим область под кривой на [0, 3]
ix = np.linspace(0, 3, 100)
iy = f(ix)
ax.fill_between(ix, iy, color='gray', alpha=0.3, label=f'Площадь = {area:.4f}')

# Добавим вертикальные линии
ax.axvline(x=0, color='gray', linestyle='--', linewidth=1)
ax.axvline(x=3, color='gray', linestyle='--', linewidth=1)

# Оформление
ax.set_title(r'График $f(x) = x^2 e^{-x}$ и площадь на $[0, 3]$')
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.legend()
ax.grid(True)

# Корректная строка заголовка с использованием f-строки
plt.title(f'$\\int_0^3 x^2 e^{{-x}}\\,dx = 2 - \\frac{{17}}{{e^3}} \\approx {area:.4f}$')

plt.show()
```

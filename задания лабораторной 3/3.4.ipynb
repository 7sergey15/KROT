```python
import numpy as np
import matplotlib.pyplot as plt

#   
def f(x):
    return x**2

a, b = 0, 3
area = (b**3)/3 - (a**3)/3  # xdx = x/3

#    
x = np.linspace(a, b, 400)
y = f(x)

#  
fig, ax = plt.subplots(figsize=(10, 6))

#     
ax.plot(x, y, 'r-', linewidth=2)
ax.fill_between(x, y, alpha=0.3, color='blue')

#  
ax.set_title(r'   $f(x) = x^2$  [0, 3]', fontsize=14)
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('f(x)', fontsize=12)
ax.grid(True, linestyle='--', alpha=0.7)

#  
ax.text(1.5, 5, f' = {area:.2f}', 
        fontsize=12, ha='center', 
        bbox=dict(facecolor='white', alpha=0.9))

#  
ax.set_xlim([a-0.2, b+0.2])
ax.set_ylim([0, 10])

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

#  
def f(x):
    return np.sin(x)

#   
a, b = 0, np.pi
analytic_area = -np.cos(b) + np.cos(a)  # sin(x)dx = -cos(x) + C

#  
numeric_area, error = quad(f, a, b)

#    
x = np.linspace(a, b, 200)
y = f(x)

#  
fig, ax = plt.subplots(figsize=(10, 6))

#     
ax.plot(x, y, 'r-', linewidth=2, label='f(x) = sin(x)')
ax.fill_between(x, y, alpha=0.3, color='blue')

#  
ax.set_title(r'   $f(x) = \sin(x)$  $[0, \pi]$', fontsize=14)
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('f(x)', fontsize=12)
ax.grid(True, linestyle='--', alpha=0.7)

#   
ax.set_xticks([0, np.pi/2, np.pi])
ax.set_xticklabels(['0', r'$\frac{\pi}{2}$', r'$\pi$'])
ax.text(np.pi/2, 0.4, f' = {analytic_area:.2f}', 
        fontsize=12, ha='center', va='center',
        bbox=dict(facecolor='white', alpha=0.9))

#   
ax.legend()
ax.set_xlim([a-0.2, b+0.2])
ax.set_ylim([0, 1.1])

plt.tight_layout()
plt.show()

#  
print(f" : {analytic_area:.4f}")
print(f" : {numeric_area:.4f}")
print(f": {error:.2e}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

#  
def f(x):
    return 2 * x + 3

#  
area_algebraic, error = quad(f, -2, 2)
area_geometric = abs(quad(f, -2, -1.5)[0]) + quad(f, -1.5, 2)[0]

#    
x = np.linspace(-2.5, 2.5, 400)
y = f(x)

#  
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(x, y, 'b', linewidth=2, label=r'$f(x) = 2x + 3$')
ax.axhline(y=0, color='black', linewidth=0.5)
ax.axvline(x=-1.5, color='gray', linestyle='--', linewidth=1, label='x = -1.5')

#    
ix1 = np.linspace(-2, -1.5, 100)
ix2 = np.linspace(-1.5, 2, 100)
iy1 = f(ix1)
iy2 = f(ix2)
ax.fill_between(ix1, iy1, color='red', alpha=0.3, label=f' : {abs(quad(f, -2, -1.5)[0]):.4f}')
ax.fill_between(ix2, iy2, color='green', alpha=0.3, label=f' : {quad(f, -1.5, 2)[0]:.4f}')

# 
ax.set_title(r' $f(x) = 2x + 3$    $[-2, 2]$')
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.legend()
ax.grid(True)

#   
plt.title(f'$\\int_{{-2}}^2 (2x + 3)\\,dx = {area_algebraic:.4f}$\n : {area_geometric:.4f}$')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

#  
def f(x):
    return x / (1 + x**2)

#  
area_algebraic, error = quad(f, -2, 2)
area_positive, error = quad(f, 0, 2)
area_geometric = 2 * area_positive

#    
x = np.linspace(-2.5, 2.5, 400)
y = f(x)

#  
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(x, y, 'b', linewidth=2, label=r'$f(x) = \frac{x}{1 + x^2}$')
ax.axhline(y=0, color='black', linewidth=0.5)

#    
ix_neg = np.linspace(-2, 0, 100)
ix_pos = np.linspace(0, 2, 100)
iy_neg = f(ix_neg)
iy_pos = f(ix_pos)
ax.fill_between(ix_neg, iy_neg, color='red', alpha=0.3, label=f' : {abs(area_positive):.4f}')
ax.fill_between(ix_pos, iy_pos, color='green', alpha=0.3, label=f' : {area_positive:.4f}')

# 
ax.set_title(r' $f(x) = \frac{x}{1 + x^2}$    $[-2, 2]$')
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.legend()
ax.grid(True)

#   
plt.title(f'$\\int_{{-2}}^2 \\frac{{x}}{{1 + x^2}}\\,dx = {area_algebraic:.4f}$\n : {area_geometric:.4f} = \\ln(5) \\approx {np.log(5):.4f}$')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Создание Series с исходными данными
series = pd.Series(
    data=[120, 130, 110, 140, 135],
    index=['2024-04-01', '2024-04-02', '2024-04-03', '2024-04-04', '2024-04-05']
)

# Преобразование индекса в DatetimeIndex
series.index = pd.to_datetime(series.index)

# Настройка графика
plt.figure(figsize=(10, 5))
plt.plot(series, marker='o', linestyle='-', color='green')

# Добавление заголовка и подписей
plt.title('Динамика значений за апрель 2024', fontsize=14)
plt.xlabel('Дата', fontsize=12)
plt.ylabel('Значение', fontsize=12)

# Форматирование дат на оси X
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()

# Отображение графика
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, OneHotEncoder
import matplotlib.pyplot as plt
import seaborn as sns

# --------------------------
# 1. Обзор структуры данных
# --------------------------
df = pd.read_csv("time_series_covid19_confirmed_global.csv")

# Проверка структуры
print("Первые 5 строк:")
display(df.head())

print("\nИнформация о данных:")
print(df.info())

print("\nОписательная статистика:")
display(df.describe(include='all'))

# Анализ целевого признака (последняя дата в данных)
target_col = df.columns[-1]
df.rename(columns={target_col: 'target'}, inplace=True)
print(f"\nЦелевой признак: {target_col}")

# --------------------------
# 2. Обработка пропусков
# --------------------------
print("\nПропуски до обработки:")
print(df.isnull().sum())

# Безопасное заполнение пропусков
df = df.assign(**{
    'Province/State': df['Province/State'].fillna('Unknown'),
    'Lat': df['Lat'].fillna(df['Lat'].median()),
    'Long': df['Long'].fillna(df['Long'].median())
})

print("\nПропуски после обработки:")
print(df.isnull().sum())

# --------------------------
# 3. Обработка выбросов (на примере координат)
# --------------------------
numerical_cols = ['Lat', 'Long']

def remove_outliers_iqr(data, columns):
    cleaned_data = data.copy()
    for col in columns:
        Q1 = data[col].quantile(0.25)
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1
        cleaned_data = cleaned_data[(cleaned_data[col] >= (Q1 - 1.5*IQR)) & 
                                  (cleaned_data[col] <= (Q3 + 1.5*IQR))]
    return cleaned_data

original_size = df.shape[0]
df_clean = remove_outliers_iqr(df, numerical_cols)
new_size = df_clean.shape[0]

print(f"\nОбработка выбросов:")
print(f"Удалено записей: {original_size - new_size} ({(original_size - new_size)/original_size*100:.2f}%)")

# --------------------------
# 4. Масштабирование числовых признаков
# --------------------------
scaler = StandardScaler()
df_clean[numerical_cols] = scaler.fit_transform(df_clean[numerical_cols])

print("\nПосле масштабирования:")
display(df_clean[numerical_cols].head())

# --------------------------
# 5. Кодирование категориальных признаков
# --------------------------
categorical_cols = ['Country/Region', 'Province/State']

encoder = OneHotEncoder(drop='first', sparse_output=False)
encoded_data = encoder.fit_transform(df_clean[categorical_cols])
encoded_df = pd.DataFrame(encoded_data, 
                        columns=encoder.get_feature_names_out(categorical_cols),
                        index=df_clean.index)

df_final = pd.concat([df_clean.drop(categorical_cols, axis=1), encoded_df], axis=1)

print("\nПроверка дамми-ловушки:")
print(f"Оригинальные категории: {len(categorical_cols)}")
print(f"Создано OHE-колонок: {encoded_df.shape[1]}")

# --------------------------
# 6. Финальный набор данных
# --------------------------
print("\nФинальная проверка:")
print(f"Размерность: {df_final.shape}")
print(f"Типы данных:\n{df_final.dtypes.value_counts()}")
print(f"Оставшиеся пропуски: {df_final.isnull().sum().sum()}")

# Визуализация распределения целевой переменной
plt.figure(figsize=(10,6))
sns.histplot(df_final['target'], kde=True, bins=50)
plt.title('Распределение подтвержденных случаев COVID-19')
plt.xlabel('Количество случаев')
plt.ylabel('Частота')
plt.show()
```
